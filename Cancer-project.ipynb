```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= np.append(cancer['feature_names'],['target']))
```

```python
correlation_matrix = df_cancer.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', fmt=".2f")
plt.title("matrice de corrélation")
plt.show()
```

```python
sns.pairplot(df_cancer ,hue ='target', vars =['mean radius','mean texture', 'mean perimeter', 'mean area',
 'mean smoothness', 'mean compactness' ,'mean concavity'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
features=['mean radius','mean texture', 'mean perimeter', 'mean area',
 'mean smoothness', 'mean compactness' ,'mean concavity']
target='target'
X = df_cancer[features]
y = df_cancer[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
classification_model = LogisticRegression()
classification_model.fit(X_train, y_train)
classification_score=classification_model.score(X_test, y_test)
classification_score
```

```python
from sklearn.metrics import confusion_matrix
y_predict = classification_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_predict)
print("Coefficient de détermination R2 :", r2)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

features_clustering=['mean smoothness','mean radius']
X_clustering = df_cancer[features_clustering]

kmeans = KMeans(n_clusters=3) 
kmeans.fit(X_clustering)

clusters = kmeans.predict(X_clustering)

silhouette_avg = silhouette_score(X_clustering, clusters)
print("Silhouette Score:", silhouette_avg)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_clustering['mean smoothness'], X_clustering['mean radius'], c=clusters)
plt.xlabel('mean smoothness')
plt.ylabel('mean radius')
plt.title('K-means Clustering')
plt.colorbar(label='Cluster')
plt.show()
```
